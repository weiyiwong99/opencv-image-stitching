```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img1 = cv2.imread("first_image.jpg")
img2 = cv2.imread("second_image.jpg")
img1_gray = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
img2_gray = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
plt.imshow(img1_gray, cmap='gray')
plt.show()
plt.imshow(img2_gray, cmap='gray')
plt.show()
```

```python
orb = cv2.ORB_create(nfeatures=2000)
keypoints1, descriptors1 = orb.detectAndCompute(img1, None)
keypoints2, descriptors2 = orb.detectAndCompute(img2, None)
plt.imshow(cv2.drawKeypoints(img1, keypoints1, None, (255, 0, 255)))
plt.show()
plt.imshow(cv2.drawKeypoints(img2, keypoints2, None, (255, 0, 255)))
plt.show()
```

```python
bf = cv2.BFMatcher_create(cv2.NORM_HAMMING)
matches = bf.knnMatch(descriptors1, descriptors2, k=2)
```

```python
def drawMatches(img1, keypoints1, img2, keypoints2, matches):
    height1, width1 = img1.shape[:2]
    height2, width2 = img2.shape[:2]
    out_img = np.zeros((max(height1, height2), width1+width2, 3), dtype='uint8') # uint8 => an int between 1-255
    out_img[:height1, :width1, :] = np.dstack((img1, img1, img1))
    out_img[:height2, width1:width1+width2, :] = np.dstack((img2, img2, img2))
    
    for match in matches:
        img1_idx = match.queryIdx
        img2_idx = match.trainIdx
        x1, y1 = keypoints1[img1_idx].pt
        x2, y2 = keypoints2[img2_idx].pt
        cv2.circle(out_img, (int(x1), int(y1)), 4, (255, 0, 255), 2)
        cv2.circle(out_img, (int(x2)+width1, int(y2)), 4, (255, 0, 255), 2)
        cv2.line(out_img, (int(x1), int(y1)), (int(x2)+width1, int(y2)), (255, 0, 255), 2)
    
    return out_img
```

```python
best_matches = []
for m, n in matches:
    best_matches.append(m)
out_img = drawMatches(img1_gray, keypoints1, img2_gray, keypoints2, best_matches)
plt.imshow(out_img)
plt.show()
```

```python
good = []
for m, n in matches:
    if m.distance < 0.6 * n.distance:
        good.append(m)
out_img = drawMatches(img1_gray, keypoints1, img2_gray, keypoints2, good)
plt.imshow(out_img)
plt.show()
```

```python
def warpImages(img1, img2, H):
    height1, width1 = img1.shape[:2]
    height2, width2 = img2.shape[:2]
    
    list_of_points_1 = np.float32([[0, 0], [0, height1], [width1, height1], [width1, 0]]).reshape(-1, 1, 2)
    temp_points = np.float32([[0, 0], [0, height2], [width2, height2], [width2, 0]]).reshape(-1, 1, 2)
    
    list_of_points_2 = cv2.perspectiveTransform(temp_points, H)
    list_of_points = np.concatenate((list_of_points_1, list_of_points_2), axis=0)
    
    [x_min, y_min] = np.int32(list_of_points.min(axis=0).ravel() - 0.5)
    [x_max, y_max] = np.int32(list_of_points.max(axis=0).ravel() + 0.5)
    
    translation_dist = [-x_min, -y_min]
    
    H_translation = np.array([[1, 0, translation_dist[0]], [0, 1, translation_dist[1]], [0, 0, 1]])
    
    out_img = cv2.warpPerspective(img2, H_translation.dot(H), (x_max-x_min, y_max-y_min))
    out_img[translation_dist[1]:height1+translation_dist[1], translation_dist[0]:width1+translation_dist[0]] = img1
    
    return out_img
```

```python
MIN_MATCH_COUNT = 10

if len(good) > MIN_MATCH_COUNT:
    src_pts = np.float32([keypoints1[m.queryIdx].pt for m in good]).reshape(-1, 1, 2)
    dst_pts = np.float32([keypoints2[m.trainIdx].pt for m in good]).reshape(-1, 1, 2)
    M, _ = cv2.findHomography(src_pts, dst_pts, cv2.RANSAC, 5.0)
    stitched_img = warpImages(img2, img1, M)
    plt.imshow(stitched_img)
    plt.savefig('stitched_img.jpg')
    plt.show()
```
